```python
def get_data(file):
  return [json.loads(line) for line in open(file, 'r')]

# one ngo test
test_data = get_data(files[0])[0]
test_data
```

Selecting only some data: name, OKVED (ОКВЭД), opf (ОКОПФ), status (acting/dead), website

```python
def get_okved_opf(nko):
  return  {'name' : nko['shortName'],
           'okved' : nko['mainOkved']['name'],
           'opf': nko['opf']['name'],
           'status' : nko['statusDetail']['shortName'],
           'website' : nko['website']}
```

```python
nkos = []

for file in files:
  d = get_data(file)
  for nko in d:
    nkos.append(get_okved_opf(nko))
```

```python
df = pd.DataFrame(nkos)
df.groupby('status').count()
```
